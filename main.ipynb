```python
import lxml
import pandas as pd
from lxml import html
import requests
```

```python
data = requests.get("https://www.iplt20.com/schedule")
```

```python
#data.content
```

```python
htmlData = html.fromstring(data.content)
```

```python
rootNode = htmlData.xpath('//div[@class="js-list"]')
```

```python
rootNode
```

```python
teamAList = []
teamBList = []
timeList = []

for element in rootNode:
    for _element in element.xpath('div'):
        teams = _element.xpath('div[@class="fixture__teams"]')
        for team in teams:
            teamA = team.xpath('div[@class="fixture__team"]/p[@class="fixture__team-name"]/text()')[0]
            teamB = team.xpath('div[@class="fixture__team"]/p[@class="fixture__team-name"]/text()')[1]
            #print(teamA.__class__)
            teamAList.append(teamA)
            teamBList.append(teamB)
            #print(teamA, ',',teamB)
            time = _element.xpath('div[@class="fixture__info"]/strong[@class="fixture__time"]/text()')
            #print(time)
            stadium = _element.xpath('div[@class="fixture__info"]/span/text()')
            #print(stadium)
        timeList.append(time)
print(teamAList, teamBList, timeList)
```

```python
dict_schedule = {'teamA': teamAList, 'teamB': teamBList, 'time': timeList}
```

```python
scheduleDF = pd.DataFrame.from_dict(dict_schedule)
scheduleDF
```
